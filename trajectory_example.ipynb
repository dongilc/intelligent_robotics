```python
import intelligent_robotics as ir
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Target DEG 20000
# VMAX 1000 DPS
# AMAX 1000
target = 10000
v = 1000/target
a = v/2

# normalized
time, traj = ir.Trapezoidal_Traj_Gen_Given_Vmax_and_Amax(v,a,0.001)
plt.plot(time, traj)
plt.legend(['s','v','a'])
plt.xlabel('time[sec]')
plt.show()

# scaled
plt.plot(time, traj*target)
plt.legend(['s','v','a'])
plt.xlabel('time[sec]')
plt.show()

# pos traj
x_traj = ir.Path_Gen(0,target,traj[:,0])
plt.plot(time, x_traj)
plt.legend(['x'])
plt.xlabel('time[sec]')
plt.show()
```

```python
desired_pos = 10
time, traj = ir.Trapezoidal_Traj_Gen_Given_Amax_and_T(0.5,3,0.001)
p_traj = ir.Path_Gen(0,desired_pos,traj[:,0])
plt.plot(time, p_traj)
plt.legend(['pos'])
plt.xlabel('time[sec]')
plt.show()
```

```python
plt.plot(time, desired_pos*traj[:,1])
plt.legend(['vel'])
plt.xlabel('time[sec]')
plt.show()
```
